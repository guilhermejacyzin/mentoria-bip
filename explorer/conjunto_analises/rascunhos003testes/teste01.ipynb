```python
# calcula a moda global da coluna "nuvem_9h", ignorando valores nulos
# essa moda global será usada como fallback quando uma localidade tiver todos os valores nulos
moda_global_nuvem_9h = df["nuvem_9h"].dropna().mode().iloc[0]
```

```python
# define uma função que recebe uma série de valores de uma localidade
# essa função retorna a moda da localidade, caso exista
# se a moda não existir porque a série está vazia (todos os valores eram nulos), retorna a moda global
def obter_moda_ou_fallback_nuvem_9h(serie_localidade):
    # calcula a moda da série da localidade, ignorando valores nulos
    moda_local = serie_localidade.dropna().mode()

    # verifica se a moda da localidade ficou vazia (isso acontece quando a localidade tem tudo nulo)
    if moda_local.empty:
        # retorna a moda global como fallback
        return moda_global_nuvem_9h

    # retorna o primeiro valor da moda local
    return moda_local.iloc[0]
```

```python
# cria uma série com a moda de "nuvem_9h" por localidade, usando a função com fallback
moda_nuvem_9h_por_localidade = (
    df
    .groupby("localidade")["nuvem_9h"]
    .apply(obter_moda_ou_fallback_nuvem_9h)
)
```

```python
# preenche os valores nulos de "nuvem_9h" usando a moda correspondente da localidade
# se a localidade não tiver moda porque era tudo nulo, o fallback garante um valor válido
df["nuvem_9h"] = df["nuvem_9h"].fillna(df["localidade"].map(moda_nuvem_9h_por_localidade))
```

```python
# define uma função que recebe uma série de valores de uma localidade
# essa função retorna a moda da localidade, caso exista
# se a moda não existir porque a série está vazia (todos os valores eram nulos), retorna a moda global
def obter_moda_ou_fallback_nuvem_15h(serie_localidade):
    # calcula a moda da série da localidade, ignorando valores nulos
    moda_local = serie_localidade.dropna().mode()

    # verifica se a moda da localidade ficou vazia (isso acontece quando a localidade tem tudo nulo)
    if moda_local.empty:
        # retorna a moda global como fallback
        return moda_global_nuvem_15h

    # retorna o primeiro valor da moda local
    return moda_local.iloc[0]
```

```python
# cria uma série com a moda de "nuvem_15h" por localidade, usando a função com fallback
moda_nuvem_15h_por_localidade = (
    df
    .groupby("localidade")["nuvem_15h"]
    .apply(obter_moda_ou_fallback_nuvem_15h)
)
```

```python
# a coluna "variacao_temp_dia" é criada no dataframe
# ela recebe a diferença entre a temperatura das 15h e a temperatura das 9h
# valores positivos indicam aquecimento ao longo do dia
df["variacao_temp_dia"] = df["temp_15h"] - df["temp_9h"]
```

```python
# a coluna "delta_nuvem" é criada no dataframe
# ela recebe a diferença entre a nebulosidade às 15h e às 9h
# aumento de nuvem ao longo do dia indica maior instabilidade
df["delta_nuvem"] = df["nuvem_15h"] - df["nuvem_9h"]
```

```python
# a coluna "queda_pressao" é criada no dataframe
# ela recebe o valor 1 quando a pressão às 15h é menor que às 9h
# quedas de pressão são indicativo clássico de instabilidade climática
df["queda_pressao"] = (df["pressao_15h"] < df["pressao_9h"]).astype(int)
```

```python
# a coluna "chuva_persistente" é criada no dataframe
# ela recebe o valor 1 quando choveu hoje e também choveu ontem
# para isso, usamos um shift de 1 dia dentro de cada localidade
df["chuva_persistente"] = (
    (df["choveu_hoje"] == 1) &
    (df.groupby("localidade")["choveu_hoje"].shift(1) == 1)
).astype(int)
```

```python
# cria uma coluna auxiliar que identifica dias sem chuva
# recebe 1 quando não choveu no dia, e 0 quando choveu
df["sem_chuva"] = (df["choveu_hoje"] == 0).astype(int)
```

```python
# a coluna "dias_sem_chuva" é criada
# ela calcula quantos dias consecutivos sem chuva existem até o dia atual
# o cálculo é feito separadamente por localidade
df["dias_sem_chuva"] = (
    df
    .groupby("localidade")["sem_chuva"]
    .cumsum()
)
```

```python
# a coluna "vento_forte" é criada
# ela recebe 1 quando a velocidade da rajada de vento ultrapassa 40 km/h
# esse limiar representa eventos potencialmente severos
df["vento_forte"] = (df["vento_rajada_velocidade"] >= 40).astype(int)
```

```python
# a coluna "calor_umido" é criada
# ela recebe 1 quando a temperatura às 15h é alta e a umidade também é alta
# essa combinação favorece instabilidade atmosférica
df["calor_umido"] = (
    (df["temp_15h"] >= 30) &
    (df["umidade_15h"] >= 70)
).astype(int)
```

```python
df_enso = pd.read_csv(r'C:\Users\JacyzinGuilherme(Bip\mentoria-bip\dados_brutos\el_nino_la_nina.txt', delim_whitespace=True)
df_enso
```

```python
# cria um dicionário que mapeia cada estação ENSO para o mês central correspondente
mapa_seas_para_mes = {
    "DJF": 1,
    "JFM": 2,
    "FMA": 3,
    "MAM": 4,
    "AMJ": 5,
    "MJJ": 6,
    "JJA": 7,
    "JAS": 8,
    "ASO": 9,
    "SON": 10,
    "OND": 11,
    "NDJ": 12
}
```

```python
# cria a coluna "mes" no dataframe ENSO
# essa coluna recebe o mês central correspondente à estação SEAS
df_enso["mes"] = df_enso["SEAS"].map(mapa_seas_para_mes)
```

```python
# ajusta o ano para a estação DJF
# quando SEAS é DJF, o ano é incrementado em 1
df_enso.loc[df_enso["SEAS"] == "DJF", "YR"] += 1
```

```python
# cria a coluna "fase_enso"
# classifica a anomalia em El Nino, La Nina ou Neutro
df_enso["fase_enso"] = np.where(
    df_enso["ANOM"] >= 0.5, "ElNino",
    np.where(df_enso["ANOM"] <= -0.5, "LaNina", "Neutro")
)
```

```python
# cria a coluna "intensidade_enso"
# classifica a intensidade do ENSO com base no valor absoluto da anomalia
# o default é definido como "Neutra" para manter o tipo string e evitar conflito de dtype
df_enso["intensidade_enso"] = np.select(
    [
        df_enso["ANOM"].abs() < 0.5,
        df_enso["ANOM"].abs().between(0.5, 1.0),
        df_enso["ANOM"].abs().between(1.0, 1.5),
        df_enso["ANOM"].abs() >= 1.5
    ],
    [
        "Neutra",
        "Fraca",
        "Moderada",
        "Forte"
    ],
    default="Neutra"
)
```

```python
# cria a coluna "enso_sinal"
# recebe 1 para El Nino, -1 para La Nina e 0 para Neutro
df_enso["enso_sinal"] = np.where(
    df_enso["fase_enso"] == "ElNino", 1,
    np.where(df_enso["fase_enso"] == "LaNina", -1, 0)
)
```

```python
# cria um dataframe ENSO final apenas com colunas necessárias
# isso evita vazamento e reduz ruído
df_enso_final = df_enso[
    ["YR", "mes", "ANOM", "fase_enso", "intensidade_enso", "enso_sinal"]
].rename(columns={"YR": "ano"})
```

```python
# realiza o merge entre o dataframe climático principal e o dataframe ENSO
# o join é feito usando as colunas "ano" e "mes"
# o tipo de join é left para preservar todas as observações climáticas
df = df.merge(
    df_enso_final,
    on=["ano", "mes"],
    how="left"
)
```

```python
# cria um dataframe com pontos costeiros de referência na Austrália
# esses pontos representam cidades costeiras principais
df_costa = pd.DataFrame({
    "lat_costa": [-33.87, -37.81, -27.47, -31.95],
    "lng_costa": [151.21, 144.96, 153.03, 115.86]
})
```

```python
#!pip install geopy

from geopy.distance import geodesic

# define uma função que calcula a menor distância até a costa
def calcular_distancia_costa(lat, lng):
    # se latitude ou longitude forem nulas, retorna NaN
    if pd.isna(lat) or pd.isna(lng):
        return np.nan
    
    # calcula a menor distância entre o ponto e os pontos costeiros
    return min(
        geodesic((lat, lng), (row.lat_costa, row.lng_costa)).km
        for _, row in df_costa.iterrows()
    )
```

```python
# cria a coluna "distancia_costa"
# cada linha recebe a distância mínima até a costa em quilômetros
df["distancia_costa"] = df.apply(
    lambda x: calcular_distancia_costa(x["lat"], x["lng"]),
    axis=1
)
```

```python
# cria a coluna "ponto_orvalho_15h"
# calcula o ponto de orvalho usando temperatura e umidade das 15h
df["ponto_orvalho_15h"] = (
    df["temp_15h"] - ((100 - df["umidade_15h"]) / 5)
)
```

```python
# cria a coluna "depressao_orvalho_15h"
# representa o quão distante o ar está da saturação
df["depressao_orvalho_15h"] = (
    df["temp_15h"] - df["ponto_orvalho_15h"]
)
```
