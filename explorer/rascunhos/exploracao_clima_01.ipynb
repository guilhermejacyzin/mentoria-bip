```python
df_clima.isna().sum()
```

```python
# Colunas num√©ricas com NaN (baseado no df.info() que voc√™ rodou)
colunas_numericas = [
    'temp_min', 'temp_max', 'chuva_mm', 'evaporacao', 
    'horas_sol', 'vento_rajada_velocidade', 'vento_veloc_9h', 
    'vento_veloc_15h', 'umidade_9h', 'umidade_15h', 
    'pressao_9h', 'pressao_15h', 'nuvem_9h', 'nuvem_15h', 
    'temp_9h', 'temp_15h'
]

print("--- Imputa√ß√£o de NaN em Colunas Num√©ricas (Mediana) ---")

for coluna in colunas_numericas:
    
    # üö® PASSO ESSENCIAL: Garante que a coluna √© num√©rica e converte texto sujo para NaN.
    # Isso resolve o TypeError.
    df_clima[coluna] = pd.to_numeric(df_clima[coluna], errors='coerce')
    
    # 1. Calcula a mediana (agora que a coluna √© garantidamente num√©rica)
    mediana = df_clima[coluna].median()
    
    # 2. Preenche os valores NaN (tanto os originais quanto os gerados pela coer√ß√£o)
    df_clima[coluna].fillna(value=mediana, inplace=True)
    
    print(f"'{coluna}' preenchida com mediana: {mediana:.2f}")
```
