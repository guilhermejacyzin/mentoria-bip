```python


fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Graf nuvem 9h
prop_9h = (df_clima.groupby(['nuvem_9h', 'chove_amanha_vtr']).size().unstack())

prop_9h['proporcao'] = prop_9h['Yes'] / (prop_9h['Yes'] + prop_9h['No'])

axes[0].bar(prop_9h.index, prop_9h['proporcao'])
axes[0].set_title('Nuvem Ã s 9h')
axes[0].set_xlabel('NÃ­vel de Nuvem')
axes[0].set_ylabel('ProporÃ§Ã£o % de Chuva AmanhÃ£')

# Graf nuvem 15h
prop_15h = (df_clima.groupby(['nuvem_15h', 'chove_amanha_vtr']).size().unstack())

prop_15h['proporcao'] = prop_15h['Yes'] / (prop_15h['Yes'] + prop_15h['No'])

axes[1].bar(prop_15h.index, prop_15h['proporcao'])
axes[1].set_title('Nuvem Ã s 15h')
axes[1].set_xlabel('NÃ­vel de Nuvem')

plt.suptitle("ProporÃ§Ã£o de Chuva AmanhÃ£ por NÃ­vel de Nuvem")
plt.tight_layout()
plt.show()
```
