```python
# 1. Carregar os datasets
df_clima = pd.read_csv(r'C:\Users\JacyzinGuilherme(Bip\mentoria-bip\dados_editados\australia_clima_v5.csv', sep=',')
df_cities = pd.read_csv(r'C:\Users\JacyzinGuilherme(Bip\mentoria-bip\dados_brutos\au_cidades.csv')
df_worldcities = pd.read_csv(r'C:\Users\JacyzinGuilherme(Bip\mentoria-bip\dados_brutos\worldcities.csv', sep=',')
df_enso = pd.read_csv(r'C:\Users\JacyzinGuilherme(Bip\mentoria-bip\dados_brutos\el_nino_la_nina.txt', sep=r'\s+')

# FunÃ§Ã£o para normalizar nomes: minÃºsculo, sem espaÃ§os nas extremidades e sem espaÃ§os duplos
def normalizar_coluna(df, nome_coluna):
    return df[nome_coluna].astype(str).str.lower().str.strip().str.replace(r'\s+', ' ', regex=True)

# Normalizar localidade no dataset principal
df_clima['localidade_match'] = normalizar_coluna(df_clima, 'localidade')

# Preparar o dataset (au_cidades)
df_cities = df_cities[['city', 'lat', 'lng']].copy()
df_cities['city_match'] = normalizar_coluna(df_cities, 'city')
# Criar versÃ£o sem espaÃ§os para casos como "BadgerysCreek"
df_cities['city_no_space'] = df_cities['city_match'].str.replace(' ', '')

# Preparar o dataset (worldcities - apenas AustrÃ¡lia)
df_world = df_worldcities[df_worldcities['country'] == 'Australia'][['city', 'lat', 'lng']].copy()
df_world['city_match'] = normalizar_coluna(df_world, 'city')
df_world['city_no_space'] = df_world['city_match'].str.replace(' ', '')

# --- MERGE EM ETAPAS (FALLBACK) ---

# Passo 1: Merge com df_cities (Dataset 2)
df_clima = pd.merge(df_clima, df_cities[['city_match', 'lat', 'lng']], 
                    left_on='localidade_match', right_on='city_match', how='left')
df_clima = df_clima.drop(columns=['city_match'])

# Passo 2: Tentar match sem espaÃ§os para os nulos (casos como "CoffsHarbour")
mask_nulo = df_clima['lat'].isna()
df_fix_space = pd.merge(df_clima[mask_nulo].drop(columns=['lat', 'lng']), 
                        df_cities[['city_no_space', 'lat', 'lng']], 
                        left_on='localidade_match', right_on='city_no_space', how='left')

# Atualizar o dataframe principal com as correÃ§Ãµes de espaÃ§o do dataset 2
df_clima.loc[mask_nulo, ['lat', 'lng']] = df_fix_space[['lat', 'lng']].values

# Passo 3: Para o que ainda estiver nulo, tentar o Dataset 3 (worldcities)
mask_ainda_nulo = df_clima['lat'].isna()
if mask_ainda_nulo.any():
    df_fix_world = pd.merge(df_clima[mask_ainda_nulo].drop(columns=['lat', 'lng']), 
                            df_world[['city_match', 'lat', 'lng']], 
                            left_on='localidade_match', right_on='city_match', how='left')
    df_clima.loc[mask_ainda_nulo, ['lat', 'lng']] = df_fix_world[['lat', 'lng']].values

# --- LIMPEZA FINAL ---

# Remover colunas auxiliares
df_clima = df_clima.drop(columns=['localidade_match'])

# Opcional: Remover linhas que ainda restaram sem coordenadas (como "MountGinini" que sÃ£o estaÃ§Ãµes)
df_final = df_clima.dropna(subset=['lat', 'lng']).reset_index(drop=True)

print(f"Total original: {len(df_clima)}")
print(f"Total com coordenadas encontradas: {len(df_final)}")
print(f"Cidades nÃ£o encontradas: {df_clima[df_clima['lat'].isna()]['localidade'].unique()}")
```

```python
# Mapeando o mÃªs central de cada temporada para garantir o merge correto
mapa_meses = {
    'DJF': 1, 'JFM': 2, 'FMA': 3, 'MAM': 4, 'AMJ': 5, 'MJJ': 6,
    'JJA': 7, 'JAS': 8, 'ASO': 9, 'SON': 10, 'OND': 11, 'NDJ': 12
}
df_enso['mes_ref'] = df_enso['SEAS'].map(mapa_meses)

# 3. Realizar o merge (Mantendo as colunas do clima originais)
df_final_2 = pd.merge(
    df_clima,
    df_enso[['YR', 'mes_ref', 'TOTAL', 'ANOM']],
    left_on=['ano', 'mes'],
    right_on=['YR', 'mes_ref'],
    how='left'
)

# 4. Criar a classificaÃ§Ã£o do evento
def classificar_status(anomalia):
    if pd.isna(anomalia): return 'sem_dados'
    if anomalia >= 0.5: return 'el_nino'
    if anomalia <= -0.5: return 'la_nina'
    return 'neutro'

df_final_2['evento_enso'] = df_final_2['ANOM'].apply(classificar_status)

# 5. RENOMEAR APENAS AS COLUNAS NOVAS para o padrÃ£o pt_br
df_final_2 = df_final_2.rename(columns={
    'ANOM': 'anomalia_enso',
    'TOTAL': 'indice_total_enso'
})
```

```python
# 1. Carregar os datasets
df_clima = pd.read_csv(r'C:\Users\JacyzinGuilherme(Bip\mentoria-bip\dados_editados\australia_clima_v5.csv', sep=',')
df_cities = pd.read_csv(r'C:\Users\JacyzinGuilherme(Bip\mentoria-bip\dados_brutos\au_cidades.csv')
df_worldcities = pd.read_csv(r'C:\Users\JacyzinGuilherme(Bip\mentoria-bip\dados_brutos\worldcities.csv', sep=',')
df_enso = pd.read_csv(r'C:\Users\JacyzinGuilherme(Bip\mentoria-bip\dados_brutos\el_nino_la_nina.txt', sep=r'\s+')

# Resolvemos os aeroportos e nomes especÃ­ficos antes do merge
mapeamento_estacoes = {
    'SydneyAirport': 'Sydney',
    'MelbourneAirport': 'Melbourne',
    'PerthAirport': 'Perth',
    'NorfolkIsland': 'Norfolk Island'
}
df_clima['localidade'] = df_clima['localidade'].replace(mapeamento_estacoes)

# FunÃ§Ã£o auxiliar para normalizar nomes
def normalizar(df, col):
    return df[col].astype(str).str.lower().str.strip().str.replace(r'\s+', ' ', regex=True)

# --- PREPARAÃ‡ÃƒO DAS FONTES DE COORDENADAS ---
df_cities_prep = df_cities[['city', 'lat', 'lng']].copy()
df_cities_prep['chave'] = normalizar(df_cities_prep, 'city')
df_cities_prep['chave_nosp'] = df_cities_prep['chave'].str.replace(' ', '')

df_world = df_worldcities[df_worldcities['country'] == 'Australia'][['city', 'lat', 'lng']].copy()
df_world['chave'] = normalizar(df_world, 'city')
df_world['chave_nosp'] = df_world['chave'].str.replace(' ', '')

# Deduplicar Ã­ndices para evitar erros de merge
cities_match = df_cities_prep.drop_duplicates('chave').set_index('chave')
cities_nosp = df_cities_prep.drop_duplicates('chave_nosp').set_index('chave_nosp')
world_match = df_world.drop_duplicates('chave').set_index('chave')
world_nosp = df_world.drop_duplicates('chave_nosp').set_index('chave_nosp')

# --- MERGE DE COORDENADAS (FALLBACK) ---
df_clima['loc_match'] = normalizar(df_clima, 'localidade')
df_clima['loc_nosp'] = df_clima['loc_match'].str.replace(' ', '')
df_clima['latitude'] = np.nan
df_clima['longitude'] = np.nan

# Passo 1: Match exato no au_cidades
df_clima['latitude'] = df_clima['latitude'].fillna(df_clima['loc_match'].map(cities_match['lat']))
df_clima['longitude'] = df_clima['longitude'].fillna(df_clima['loc_match'].map(cities_match['lng']))

# Passo 2: Match sem espaÃ§os no au_cidades (ex: AliceSprings)
mask = df_clima['latitude'].isna()
df_clima.loc[mask, 'latitude'] = df_clima.loc[mask, 'latitude'].fillna(df_clima['loc_nosp'].map(cities_nosp['lat']))
df_clima.loc[mask, 'longitude'] = df_clima.loc[mask, 'longitude'].fillna(df_clima['loc_nosp'].map(cities_nosp['lng']))

# Passo 3: Match exato no worldcities
mask = df_clima['latitude'].isna()
df_clima.loc[mask, 'latitude'] = df_clima.loc[mask, 'latitude'].fillna(df_clima['loc_match'].map(world_match['lat']))
df_clima.loc[mask, 'longitude'] = df_clima.loc[mask, 'longitude'].fillna(df_clima['loc_match'].map(world_match['lng']))

# Passo 4: Match sem espaÃ§os no worldcities
mask = df_clima['latitude'].isna()
df_clima.loc[mask, 'latitude'] = df_clima.loc[mask, 'latitude'].fillna(df_clima['loc_nosp'].map(world_nosp['lat']))
df_clima.loc[mask, 'longitude'] = df_clima.loc[mask, 'longitude'].fillna(df_clima['loc_nosp'].map(world_nosp['lng']))

# --- MERGE ENSO (EL NIÃ‘O / LA NIÃ‘A) ---
mapa_meses = {'DJF': 1, 'JFM': 2, 'FMA': 3, 'MAM': 4, 'AMJ': 5, 'MJJ': 6, 
              'JJA': 7, 'JAS': 8, 'ASO': 9, 'SON': 10, 'OND': 11, 'NDJ': 12}
df_enso['mes_ref'] = df_enso['SEAS'].map(mapa_meses)

df_clima = pd.merge(df_clima, df_enso[['YR', 'mes_ref', 'TOTAL', 'ANOM']], 
                    left_on=['ano', 'mes'], right_on=['YR', 'mes_ref'], how='left')

# ClassificaÃ§Ã£o e RenomeaÃ§Ã£o para padrÃ£o pt_br apenas nas novas
df_clima['evento_enso'] = df_clima['ANOM'].apply(
    lambda x: 'el_nino' if x >= 0.5 else ('la_nina' if x <= -0.5 else 'neutro' if pd.notna(x) else 'sem_dados')
)

df_clima = df_clima.rename(columns={
    'ANOM': 'anomalia_enso',
    'TOTAL': 'indice_total_enso'
})

# Limpeza de colunas temporÃ¡rias
df_clima = df_clima.drop(columns=['loc_match', 'loc_nosp', 'YR', 'mes_ref'], errors='ignore')

# Salvar arquivo final
df_clima.to_csv('australia_clima_final.csv', index=False)

# ValidaÃ§Ã£o rÃ¡pida
print(f"Processo concluÃ­do.")
print(f"Coordenadas para AliceSprings: {df_clima[df_clima['localidade'] == 'AliceSprings'][['latitude', 'longitude']].iloc[0].values}")
print(f"Coordenadas para Sydney (ex-Airport): {df_clima[df_clima['localidade'] == 'Sydney'][['latitude', 'longitude']].iloc[0].values}")
print(f"Cidades ainda sem coordenadas: {df_clima[df_clima['latitude'].isna()]['localidade'].unique()}")
```
