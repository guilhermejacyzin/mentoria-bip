```python
df_clima['data'] = pd.to_datetime(
    df_clima['data'].astype(str).str.strip(),
    dayfirst=True,      # DIA primeiro
    errors='coerce'     # datas estranhas viram NaT em vez de dar erro
)

df_clima['ano'] = df_clima['data'].dt.year
df_clima['mes'] = df_clima['data'].dt.month
```

```python
sazonalidade_clima = (
    df_clima
    .groupby(['localidade', 'mes'])
    .agg(
        temp_min_media = ('temp_min', 'mean'),
        temp_max_media = ('temp_max', 'mean'),
        chuva_media = ('chuva_mm', 'mean'),
        dias_chuva_pct = ('chove_amanha', lambda x: (x == 'Yes').mean() * 100)
    )
    .reset_index()
)

display(round(sazonalidade_clima, 2))
```

```python
# vendas médias por cidade
vendas_cidade = (
    df_retail.groupby('localidade')['vendas_medias_diarias']
    .mean()
    .rename('vendas_media')
)

# junta no dataset de sazonalidade
sazonalidade_merge = sazonalidade_clima.merge(
    vendas_cidade,
    on='localidade',
    how='left'
)

display(round(sazonalidade_merge, 2))
```

```python
mapa_clima = {
    'Melbourne': 'frio',
    'Adelaide': 'moderado',
    'Sydney': 'moderado',
    'Brisbane': 'quente',
    'Perth': 'quente'
}

df_retail['clima_cat'] = df_retail['localidade'].map(mapa_clima)

lucratividade_por_clima = (
    df_retail
    .groupby(['clima_cat', 'lucratividade_loja'])
    .size()
    .reset_index(name='qtd')
)
display(lucratividade_por_clima)

# total de lojas por clima
total_por_clima = lucratividade_por_clima.groupby('clima_cat')['qtd'].transform('sum')

# adicionar probabilidade
lucratividade_por_clima['probabilidade'] = (
    lucratividade_por_clima['qtd'] / total_por_clima * 100
).round(2)

display(lucratividade_por_clima)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.barplot(
    data=lucratividade_por_clima,
    x='clima_cat',
    y='probabilidade',
    hue='lucratividade_loja',
    palette='viridis'
)
plt.title('Probabilidade de Lucratividade por Tipo de Clima (%)')
plt.ylabel('Probabilidade (%)')
plt.xlabel('Clima')
plt.show()
```

```python
# 1) Resumo clima + vendas
clima_resumo = df_clima.groupby('localidade').agg(
    temp_media=('temp_max', 'mean'),
    chuva_media=('chuva_mm', 'mean'),
    umidade_media=('umidade_15h', 'mean')
)

vendas_resumo = (
    df_retail
    .groupby('localidade')['vendas_medias_diarias']
    .mean()
    .rename('vendas_media')
)

cluster_base = pd.concat([clima_resumo, vendas_resumo], axis=1)
display(round(cluster_base, 2))

# 2) "Normalizar" na mão (z-score)
cluster_norm = (cluster_base - cluster_base.mean()) / cluster_base.std()
display(cluster_norm)

# 3) Matriz de similaridade entre cidades
dist = cluster_norm.T.corr()
display(round(dist, 2))
```
