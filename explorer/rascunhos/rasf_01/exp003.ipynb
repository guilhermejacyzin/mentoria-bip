```python
# Imports

import pandas as pd
import numpy as np
```

```python
# Import da base clima para exploração
df_clima = pd.read_csv(r'C:\Users\JacyzinGuilherme(Bip\mentoria-bip\dados_brutos\clima_australia.csv', sep=';')
display(df_clima)
```

```python
df_clima = df_clima.rename(columns={'chove_amanha': 'chove_amanha_vtr', 'choveu_hoje': 'choveu_hoje_fex'})
df_clima['choveu_hoje_Nfex'] = np.where(df_clima['choveu_hoje_fex'] == 'No', 0, 1)
df_clima['chove_amanha_Nvtr'] = np.where(df_clima['chove_amanha_vtr'] == 'No', 0, 1)
```

```python
# Separando e tratando data

# converte a coluna data para datetime - sinalizando que o dia é o primeiro elemento da coluna
df_clima['data'] = pd.to_datetime(df_clima['data'], dayfirst=True, 
    # se a conversão falhar, retorna NaT (ou missing values)
    errors='coerce')

# criando uma coluna ano e extraindo o valor para salvar na coluna
df_clima['ano'] = df_clima['data'].dt.year

# criando uma coluna mês e extraindo o valor para salvar na coluna
df_clima['mes'] = df_clima['data'].dt.month

# criando uma coluna dia e extraindo o valor para salvar na coluna
df_clima['dia'] = df_clima['data'].dt.day

# criando uma coluna trimestre e extraindo o valor para salvar na coluna
df_clima['trimestre_quarter'] = df_clima['data'].dt.quarter

display(df_clima), df_clima.isna().sum()
```

```python
# Verificação da distribuição geral da var chove_amanha_Nvtr
display(round(df_clima['chove_amanha_vtr'].value_counts(normalize=True) * 100, 2))
```

```python
# Verificação da distribuição da var chove_amanha_vtr por trimestre
display(round(
    df_clima.groupby('trimestre_quarter')['chove_amanha_vtr'].value_counts(normalize=True).mul(100).rename('percentual').reset_index()
    ,2))
```

```python
# Distribuição da variável target chove_amanha_vtr por localidade
display(round(
   df_clima.groupby('localidade')['chove_amanha_vtr'].value_counts(normalize=True).mul(100).rename('percentual').reset_index().sort_values(by='percentual', ascending=False).head(10)
   ,2))
```

```python
# Choveu hoje vs Chove amanhã
display(round(
    df_clima.groupby('choveu_hoje_fex')['chove_amanha_vtr'].value_counts(normalize=True).mul(100).rename('percentual').reset_index()
    ,2))
```

```python
# Agrupando colunas com menos valores NaN
grupo_min_nan = df_clima[['temp_min', 'temp_max', 'chuva_mm', 'vento_veloc_9h', 'vento_veloc_15h', 'umidade_9h', 'temp_9h', 'temp_15h']]
```

```python
# Copy do df original // backup
df_clima_backup = df_clima.copy()
```

```python
# Media simples no grupo_min_nan

# para cada coluna in grupo_min_nan
for col in grupo_min_nan:
    # mediana recebe = df_clima_backup[coluna] com a mediana
    mediana = df_clima_backup[col].median()
    # df_clima_backup[coluna] recebe df_clima_backup[coluna] com NaN preenchidos com a mediana
    df_clima_backup[col] = df_clima_backup[col].fillna(mediana)
```

```python
# Choveu_hoje_fex

gp_var_categorica = ['choveu_hoje_fex']

# para cada coluna in gp_var_categoria
for col in gp_var_categorica:
    # moda recebe = resultado de df_clima_backup[col] com a moda
    moda = df_clima_backup[col].mode()[0]
    # df_clima_backup[coluna] recebe df_clima_backup[coluna] com NaN preenchidos com a moda
    df_clima_backup[col] = df_clima_backup[col].fillna(moda)
```
