```python
# Cidades com maior porcentagem de dias secos, tÃªm lojas com maior venda diÃ¡ria?


# Calcular total de dias por localidade
total_dias = (df_clima.groupby('localidade').size().rename('total_dias'))

# Dias secos por localidade // cidade
dias_secos = (df_clima[df_clima['cluster_chuva_mm'] == 'seco'].groupby('localidade').size().rename('dias_secos'))

# Merge // join dos dados - juntando total + dias secos
clima_seco = total_dias.to_frame().join(dias_secos, how='left')
clima_seco['dias_secos'] = clima_seco['dias_secos'].fillna(0)

# % dias secos
clima_seco['porcentagem_dias_secos'] = clima_seco['dias_secos'] / clima_seco['total_dias']


# Usando retail por localidade
vendas_por_cidade = df_retail.groupby('localidade')['vendas_medias_diarias'].mean().rename('vendas_medias_diarias_por_loja')

# Clima e Vendas
analise1 = clima_seco.join(vendas_por_cidade, how='left').round(2)
display(analise1)
```

```python
# Cidades com mais dias de chuva forte tem menor % de vendas?
dias_extremos = (df_clima[df_clima['cluster_chuva_mm'].isin(['Forte', 'extrema'])].groupby('localidade').size().rename('dias_chuva_forte_extrema'))

clima_extremo = total_dias.to_frame().join(dias_extremos, how='left')
clima_extremo['porcentagem_dias_chuva_forte_extrema'] = (clima_extremo['dias_chuva_forte_extrema'] / clima_extremo['total_dias'] * 100)

# Retail % Lojas lucrativas por cidade
df_retail['loja_lucrativa'] = df_retail['lucratividade_loja'].eq('Increased').astype(int)

porcentagem_lucrativas = (df_retail.groupby('localidade')['loja_lucrativa'].mean() * 100).rename('porcentagem_lojas_lucrativas')

# Juntando as tabelas
analise2 = clima_extremo.join(porcentagem_lucrativas, how='left').round(2)
display(analise2)
```

```python
# Cidades com dias mais quentes, algum negÃ³cio tem vendas mÃ©dias mais altas?
dias_quentes = (df_clima[df_clima['cluster_temp_max'] == 'quente'].groupby('localidade').size().rename('dias_quentes'))

clima_quente = total_dias.to_frame().join(dias_quentes, how='left')
clima_quente['dias_quentes'] = clima_quente['dias_quentes'].fillna(0)

clima_quente['porcentagem_dias_quentes'] = (clima_quente['dias_quentes'] / clima_quente['total_dias'] * 100)

# retail usando vendas por tipo de negÃ³cio e cidade
vendas_tipo_cidade = (df_retail.groupby(['localidade', 'tipo_negocio'])['vendas_medias_diarias'].mean().rename('vendas_medias_tipo_cidade').reset_index())

# adc % dias quentes a cada cidade
clima_quente_reset = clima_quente[['porcentagem_dias_quentes']].reset_index()

# anÃ¡lise
analise3 = vendas_tipo_cidade.merge(clima_quente_reset, on='localidade', how='left')
display(analise3.round(2))

```

```python
# Localidades com mais horas de sol em mÃ©dia tem maior satisfaÃ§Ã£o do cliente nas lojas?
horas_sol_media = (df_clima.groupby('localidade')['horas_sol'].mean().rename('horas_sol_media'))

# numerando mapa de satisfaÃ§Ã£o
mapa_satisfacao = {'Bad': 1, 'Good': 2}
df_retail['satisfacao_clientes_nota'] = df_retail['satisfacao_clientes'].map(mapa_satisfacao)

# Retail - satisfaÃ§Ã£o mÃ©dia de clientes por localidade
satisfacao_media = (df_retail.groupby('localidade')['satisfacao_clientes_nota'].mean().rename('satisfacao_clientes_media'))

# join
analise4 = (horas_sol_media.to_frame().join(satisfacao_media, how='left').round(2))

display(analise4)
```
