```python
# Imports

import pandas as pd
import numpy as np
```

```python
# Import da base clima para exploração
df_clima = pd.read_csv(r'C:\Users\JacyzinGuilherme(Bip\mentoria-bip\dados_editados\australia_clima_02.csv', sep=',')
display(df_clima)
```

```python
# Tamanho do dataset
df_clima.shape
```

```python
# Info do dataset
df_clima.info()
```

```python
# Verificação de nulos
df_clima.isnull().sum()
```

```python
# Separando e tratando data

# converte a coluna data para datetime - sinalizando que o dia é o primeiro elemento da coluna
df_clima['data'] = pd.to_datetime(df_clima['data'], dayfirst=True, 
    # se a conversão falhar, retorna NaT (ou missing values)
    errors='coerce')

# criando uma coluna ano e extraindo o valor para salvar na coluna
df_clima['ano'] = df_clima['data'].dt.year

# criando uma coluna mês e extraindo o valor para salvar na coluna
df_clima['mes'] = df_clima['data'].dt.month

# criando uma coluna dia e extraindo o valor para salvar na coluna
df_clima['dia'] = df_clima['data'].dt.day

display(df_clima), df_clima.isna().sum()
```

```python
# Verificando estatísticas gerais das colunas
df_clima.describe()
```

```python
# Tratando/separando colunas em clusters
df_clima['cluster_temp_min'] = np.where(df_clima['temp_min'] <=10, 'frio', np.where((df_clima['temp_min'] > 10) & (df_clima['temp_min'] <= 20), 'ameno', 'quente'))
df_clima['cluster_temp_max'] = np.where(df_clima['temp_max'] <=10, 'frio', np.where((df_clima['temp_max'] > 10) & (df_clima['temp_max'] <= 20), 'ameno', 'quente'))
df_clima['cluster_temp_9h'] = np.where(df_clima['temp_9h'] <=10, 'frio', np.where((df_clima['temp_9h'] > 10) & (df_clima['temp_9h'] <= 20), 'ameno', 'quente'))
df_clima['cluster_temp_15h'] = np.where(df_clima['temp_15h'] <=10, 'frio', np.where((df_clima['temp_15h'] > 10) & (df_clima['temp_15h'] <= 20), 'ameno', 'quente'))
df_clima['cluster_umidade_9h'] = np.where(df_clima['umidade_9h'] <=40, 'seca', np.where((df_clima['umidade_9h'] > 40) & (df_clima['umidade_9h'] <= 70), 'ameno', 'tropical'))
df_clima['cluster_umidade_15h'] = np.where(df_clima['umidade_15h'] <=40, 'seca', np.where((df_clima['umidade_15h'] > 40) & (df_clima['umidade_15h'] <= 70), 'ameno', 'tropical'))

df_clima['cluster_chuva_mm'] = np.where(df_clima['chuva_mm'] <= 2, 'seco',
                                        np.where((df_clima['chuva_mm'] > 2) & (df_clima['chuva_mm'] <= 10), 'moderada',
                                        np.where((df_clima['chuva_mm'] > 10) & (df_clima['chuva_mm'] <= 30), 'Forte', 'extrema')))

df_clima['cluster_rajada_vento'] = np.where(df_clima['vento_rajada_velocidade'] <= 19, 'leve',
                                        np.where((df_clima['vento_rajada_velocidade'] > 19) & (df_clima['vento_rajada_velocidade'] <= 39), 'moderado',
                                        np.where((df_clima['vento_rajada_velocidade'] > 39) & (df_clima['vento_rajada_velocidade'] <=69), 'forte', 'tempestade')))

df_clima['cluster_hora_sol']  = np.where(df_clima['horas_sol'] <= 10, 'frio extremo',
                                         np.where((df_clima['horas_sol'] > 10) & (df_clima['horas_sol'] <= 20), 'moderado - levemente ensolarado',
                                         np.where((df_clima['horas_sol'] > 20) & (df_clima['horas_sol'] <= 30), 'ensolarado', 'calor extremo')))

df_clima['cluster_pressao_at_9h'] = np.where(df_clima['pressao_9h'] <= 995, 'muito baixa',
                                            np.where((df_clima['pressao_9h'] > 995 ) & (df_clima['pressao_9h'] <= 1009), 'baixa',
                                            np.where((df_clima['pressao_9h'] > 1009) & (df_clima['pressao_9h'] <= 1019), 'normal', 'alta')))
                                            
df_clima['cluster_pressao_at_15h'] = np.where(df_clima['pressao_15h'] <= 995, 'muito baixa',
                                            np.where((df_clima['pressao_15h'] > 995 ) & (df_clima['pressao_15h'] <= 1009), 'baixa',
                                            np.where((df_clima['pressao_15h'] > 1009) & (df_clima['pressao_15h'] <= 1019), 'normal', 'alta')))

df_clima['cluster_nuvem_9h'] = np.where(df_clima['nuvem_9h'] == 0, 'céu limpo',
                                            np.where((df_clima['nuvem_9h'] > 0 ) & (df_clima['nuvem_9h'] <= 7), 'parcialmente nublado', 'nublado'))

df_clima['cluster_nuvem_9h'] = np.where(df_clima['nuvem_15h'] == 0, 'céu limpo',
                                            np.where((df_clima['nuvem_15h'] > 0 ) & (df_clima['nuvem_15h'] <= 7), 'parcialmente nublado', 'nublado'))                                            

display(df_clima)
```
