```python
# Imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_clima = pd.read_csv(r'C:\Users\JacyzinGuilherme(Bip\mentoria-bip\dados_editados\australia_clima.csv', decimal=',')
display(df_clima)
```

```python
df_clima.isna().sum()
```

```python
# =colunas numéricas com NaN 
colunas_numericas = [
    'temp_min', 'temp_max', 'chuva_mm', 'evaporacao', 
    'horas_sol', 'vento_rajada_velocidade', 'vento_veloc_9h', 
    'vento_veloc_15h', 'umidade_9h', 'umidade_15h', 
    'pressao_9h', 'pressao_15h', 'nuvem_9h', 'nuvem_15h', 
    'temp_9h', 'temp_15h'
]



for coluna in colunas_numericas:
    

    df_clima[coluna] = pd.to_numeric(df_clima[coluna], errors='coerce')
    
    # calculando mediana
    media = df_clima[coluna].mean()
    
    # Preenchendo NaN com média
    df_clima[coluna].fillna(value=media, inplace=True)
    
   
```
