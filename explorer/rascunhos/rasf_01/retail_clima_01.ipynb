```python
# Imports

import pandas as pd
import numpy as np
```

```python
# Base de dados retail

df_retail = pd.read_excel(r'C:\Users\JacyzinGuilherme(Bip\mentoria-bip\dados_editados\australia_retail.xlsx')
```

```python
display(df_retail)
```

```python
# Quais cidades têm maior venda média por loja?
localidade_maior_media_vendas_diaria = df_retail.groupby('localidade')['vendas_medias_diarias'].mean().sort_values(ascending=False)
display(round(localidade_maior_media_vendas_diaria,2))
```

```python
# Quais cidades têm maior venda média por loja?
negocios_maior_media_vendas_diaria = df_retail.groupby('tipo_negocio')['vendas_medias_diarias'].mean().sort_values(ascending=False)
display(round(negocios_maior_media_vendas_diaria,2))
```

```python
# Qual o segmento mais "forte" por região?
segmento_forte_regiao = df_retail.groupby(['localidade','tipo_negocio'], as_index=False)['vendas_medias_diarias'].mean()
segmento_forte_regiao = segmento_forte_regiao.sort_values(by='vendas_medias_diarias', ascending=False)
display(round(segmento_forte_regiao,2))
```
