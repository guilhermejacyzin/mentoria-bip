```python
# Import dataset
df_retail = pd.read_excel(r'C:\Users\JacyzinGuilherme(Bip\mentoria-bip\dados_editados\australia_retail.xlsx')
display(df_retail)
```

```python
# Total de média vendas por tipo e por loja
venda_total_diaria_por_tipo_marca = df_retail[['tipo_negocio', 'loja_negocio', 'vendas_medias_diarias']]
venda_total_diaria_por_tipo_marca = venda_total_diaria_por_tipo_marca.groupby(['tipo_negocio', 'loja_negocio'], as_index=False).sum().sort_values(by='vendas_medias_diarias', ascending=False)
display(venda_total_diaria_por_tipo_marca)
```

```python
# Desempenho médio por tipo e por loja
venda_media_diaria_por_tipo = df_retail[['tipo_negocio', 'loja_negocio', 'vendas_medias_diarias']]
venda_media_diaria_por_tipo = venda_media_diaria_por_tipo.groupby(['tipo_negocio', 'loja_negocio'], as_index=False).mean('vendas_medias_diarias').sort_values(by='vendas_medias_diarias', ascending=False)
display(round(venda_media_diaria_por_tipo, 2))
```

```python
venda_tipo = (
    df_retail[['tipo_negocio', 'vendas_medias_diarias']]
      .groupby('tipo_negocio', as_index=False)
      .agg(
          lojas=('vendas_medias_diarias', 'count'),
          venda_media_diaria=('vendas_medias_diarias', 'mean')
      )
      .sort_values(by='venda_media_diaria', ascending=False)
)
venda_tipo
```

```python
# 1) Focar só nas colunas que vamos usar
cols_satisfacao = ['satisfacao_clientes', 'vendas_medias_diarias', 'lucratividade_loja']
df_satisfacao = df_retail[cols_satisfacao].copy()

# 2) Agrupar por satisfação do cliente e calcular:
#    - média de vendas diárias
#    - quantidade de lojas
satisfacao_vendas = (
    df_satisfacao
        .groupby('satisfacao_clientes', as_index=False)
        .agg(
            qtd_lojas=('vendas_medias_diarias', 'count'),
            venda_media_diaria=('vendas_medias_diarias', 'mean')
        )
        .sort_values(by='venda_media_diaria', ascending=False)
)

# 3) Arredondar para ficar mais apresentável
satisfacao_vendas['venda_media_diaria'] = satisfacao_vendas['venda_media_diaria'].round(2)

display(satisfacao_vendas)
```

```python
# 4) Tabela de contingência: satisfação x lucratividade
tabela_satisfacao_lucro = (
    df_satisfacao
        .groupby(['satisfacao_clientes', 'lucratividade_loja'], as_index=False)
        .size()
        .rename(columns={'size': 'qtd_lojas'})
)

display(tabela_satisfacao_lucro)
```

```python
# 5) Proporção de lojas Increased/Decreased dentro de cada nível de satisfação
tabela_satisfacao_lucro_prop = (
    tabela_satisfacao_lucro
        .groupby('satisfacao_clientes')
        .apply(lambda g: g.assign(proporcao = g['qtd_lojas'] / g['qtd_lojas'].sum()))
        .reset_index(drop=True)
)

# Arredonda
tabela_satisfacao_lucro_prop['proporcao'] = (tabela_satisfacao_lucro_prop['proporcao'] * 100).round(1)

display(tabela_satisfacao_lucro_prop)
```
